```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
plt.plot(x,y,'b.-')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'rx')

plt.subplot(1,2,2)
plt.plot(y,x)
plt.xlabel("x2")
plt.ylabel("y2")
```

```python
## OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.2,0.4,0.9])
axes2 = fig.add_axes([0.6,0.2,0.9,0.9])
axes3 = fig.add_axes([0.7,0.8,0.2,0.2])


axes.plot(x,y)
axes.set_xlabel("x")
axes2.plot(y,x)
axes3.plot(x,y)
```

```python
plt.subplots(nrows=2,ncols=2)
#plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
#fig, axes = plt.subplots(nrows=2, ncols=2)

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the figure object    
#fig
```
