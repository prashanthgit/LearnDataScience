```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.arange(0,10,1)
y=np.arange(0,10,1)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x*x,y)
plt.subplot(2,2,3)
plt.plot(x,y*y)
plt.subplot(2,2,4)
plt.plot(x,y)
```
