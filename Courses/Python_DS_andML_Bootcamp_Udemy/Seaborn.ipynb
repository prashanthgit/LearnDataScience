# Matrix plots 

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tips.shape
```

```python
flights.shape
```

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
tc2 = flights.corr()

tc2
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,linewidths=1,cmap='coolwarm',linecolor='black')
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

```python
iris['species'].unique()
```

```python
 sns.pairplot(iris)
```

```python
from matplotlib import pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
