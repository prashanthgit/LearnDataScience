```python
# number of days per month
month_days = [0,31,28,31,30,31,30,31,31,30,31,30,31]

def is_leap(year:int)-> bool:
    ''' Find is Leap or not
    
    Parameter
    --------
    int year
    
    return
    -----
    True if leap, False if not leap
    '''
    return year % 4 == 0 and (year% 100 != 0 or year % 400 == 0)
    
def days_in_month(year: int, month: int)->bool:
    '''
    Rretune number of days in the month
   
    Parameters
    ---------
    1st year in four digits
    2nd month 1 to 12
    
    Return
    ------
    Number of days in year and month <int>
    or Str if invalid
    '''
    if not 1<= month <= 12:
        return 'Invalid month'
    if month == 2 and is_leap(year):
        return 29
    return month_days[month]

    
    
    
```
